```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('Corelation.csv')

# Drop OBJECTID as it is an identifier
df_features = df.drop(columns=['OBJECTID'])

# Calculate Spearman correlation matrix
spearman_corr = df_features.corr(method='spearman')

# Save to CSV
spearman_corr.to_csv('spearman_correlation_matrix.csv')

# Print the matrix
print(spearman_corr)

# Create a heatmap
plt.subplots(figsize=(10, 8))
sns.heatmap(spearman_corr, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={'shrink': .8})
plt.title('Spearman Correlation Matrix')
plt.tight_layout()
plt.savefig('spearman_correlation_heatmap.png')
```
